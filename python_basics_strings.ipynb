#in keyword
The in keyword is used to check if a value is present in a sequence (list, range, string etc.).

The in keyword is also used to iterate through a sequence in a for loop:

```
if 'sub' in 'sub string':
  print('substring in string')
```

#is keyword
The is keyword is used to check if the variables have the same reference

```
# a function to change domain name 
# of email id of people in an organistaion
def replace_domain(email,old_domain,new_domain):
  if '@'+old_domain in email:
    index = email.index('@'+old_domain)
    new_email = email[:index] + '@' + new_domain
    return new_email
  return email
```

#Formatting strings and f strings
[py docs](https://docs.python.org/3/reference/lexical_analysis.html#f-strings)

```
name = 'Sam'
age = 16
print('My name is {} , I am {} years old'.format(name,age))
print('My name is {name} , I am {age} years old'.format(age=age,name=name))
```

```
#formatting expression
price = 7.512
with_tax = 8.714
print('Price :{:.2f}, with tax: {:.2f}'.format(price,with_tax))
# {:.2f} -> formatting exp
```

```
names = ['General','Admiral','Marshal','Brigadier','Major','Commander']
scores = [float(x) for x in input('Enter spaced scores: ').split()]
for i in range(len(names)):
  print('{:>3} | {:>6.2f}'.format(names[i],scores[i]))
```

f strings take information from current context

```
country =  'U.S.A' ; size = 3;
f_string = f'{country:} is the {size:} largest country in the world';
print(f_string);
```
